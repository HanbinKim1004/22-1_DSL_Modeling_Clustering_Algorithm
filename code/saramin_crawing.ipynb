```python
#크롤링 할 대상 url 지정
url = 'https://www.saramin.co.kr/zf_user/company-search?searchWord=&area=&industry=&welfare=corp486&scale=&listingType=kosdaq%2Ckospi%2Ckonex&companyType=&employees=&revenue=&operatingRevenue=&netRevenue=&establishment=&salary=&startingSalary=&order=view&recruitCheck=&page=2'
#크롤링 대상 url과 headers를 기반으로 정보 요청
res = requests.get(url, headers = headers)

#정상적으로 정보를 받은 경우 지속, 그렇지 않은 경우 중지
res.raise_for_status()

#html문서를 beautifulsoup의 lxml 형태로 변환
soup = BeautifulSoup(res.text, 'lxml') 
```

```python
#html문서에서 cluster_text_headline nclicks(cls_pol.clsart)라는 이름의 class를 지닌 a 태그의 element들 모두 선택
a1 = soup.findAll('div', attrs = {'class' : 'wrap_list'})

#선택한 element들 내용 출력
a1
```

```python
#html문서에서 cluster_text_headline nclicks(cls_pol.clsart)라는 이름의 class를 지닌 a 태그의 element들 모두 선택
a1 = soup.findAll('strong', attrs = {'class' : 'title'})

#선택한 element들 내용 출력
a1
```

```python
#각 뉴스 제목과 링크를 추출할 반복문 생성
for li in a1:
    
    #a 태그의 element의 text 선택 및 공백 제거
    title = li.get_text()
    
    #a 태그의 element 내에서 href class attribute 정보 추출 및 링크 생성
    #link = a['href']
    
    #기사 제목과 해당 링크 출력
    print(f'a: {title}') 
    #print(f'(링크 : {link})')
```

```python
#페이지 별로 추출하는 반복문 생성
'''통근버스 운행'''

title_list = []

for i in range(1,57):
    #페이지 별로 요청을 보내고 soup의 lxml 형태로 변환
    url = f'https://www.saramin.co.kr/zf_user/company-search?searchWord=&area=&industry=&welfare=corp414&scale=&listingType=kosdaq%2Ckospi%2Ckonex&companyType=&employees=&revenue=&operatingRevenue=&netRevenue=&establishment=&salary=&startingSalary=&order=&recruitCheck=&page={i}'
    res = requests.get(url, headers = headers)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, 'lxml') 
    
    #html문서에서 cluster_text_headline nclicks(cls_pol.clsart)라는 이름의 class를 지닌 a 태그의 element들 모두 선택
    a2 = soup.findAll('strong', attrs = {'class' : 'title'})
    
    #각 뉴스 제목과 링크를 추출할 반복문 생성
    j = 1
    for a in a2:
        if j % 12 == 1 or j % 12 == 2:
            j += 1
            continue
            
        title = a.get_text().strip()
        title_list.append(title)
        j += 1
```
